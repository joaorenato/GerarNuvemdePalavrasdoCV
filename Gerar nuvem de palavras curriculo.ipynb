# Código para gerar nuvem de palavras a partir do arquivo em pdf do currículo

## Importando as bibliotecas necessárias

```python
#! pip3 install pypdf2
# importa as bibliotecas necessárias
import PyPDF2
import re
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from PIL import Image

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

## Lendo o arquivo em pdf do currículo

```python
pdf_file = open('JOÃO RENATO FALCÃO.pdf', 'rb')
#pdf_file
```

## Verificando o número de páginas do pdf

```python
read_pdf = PyPDF2.PdfFileReader(pdf_file) # Faz a leitura usando a biblioteca PyPDF2

num_paginas = read_pdf.getNumPages() # pega o numero de páginas
num_paginas
```

## Fazendo a leitura do conteúdo do PDF

```python
conteudo = '' # criando uma string vazia para o for loop

for i in range(int(num_paginas)):  
    pag = read_pdf.getPage(i)         # lendo página por página e
    conteudo += pag.extractText()     # inserindo seu conteúdo em um objeto page_content

print(conteudo)
parsed = ''.join(conteudo) # faz a junção das linhas

print("Sem eliminar as quebras")
print(parsed)


print("\nAPÓS ELIMINAR AS QUEBRAS:\n")
parsed = re.sub('\n', '', parsed) # remove as quebras de linha
print(parsed)

print("\nAPÓS ELIMINAR OS UNDERLINES\n:")
parsed = re.sub('_', '', parsed) # remove os underlines
print(parsed)

```

## Vejamos passo a passo como fica o objeto resultante da leitura do pdf

### Imprimindo o objeto a partir da primeira leitura

```python
print(conteudo)
```

## Concatenando as palavras que foram "quebradas"

```python
parsed = ''.join(conteudo) # concatenar palavras que estão quebradas
print(parsed)
```

## Vamos agora eliminar as quebras de linha do texto

```python
print("APÓS ELIMINAR AS QUEBRAS:\n")
parsed = re.sub('\n', '', parsed) # remove as quebras de linha
print(parsed)
```

Perceba que, por conta do modelo utilizado para o currículo, há vários traços ("uderlines")


## Vamos eliminar também essas linhas do nosso objeto de interesse

```python
print("\nAPÓS ELIMINAR OS UNDERLINES\n:")
parsed = re.sub('_', '', parsed)
print(parsed)
```

## Após fazer as correções necessárias, vamos criar o que é conhecido como *stopwords*

O <b>stopwords</b> é um comando da função WorldCloud cujo objetivo é "limpar" o texto a ser utilizado.

A ideia é remover palavras que não agregam ao conteúdo exibido na nuvem. Por exemplo, "Página", "Atualmente", "da", "do", etc.

Aqui, é criada uma lista com todas as palavras a serem ignoradas na criação da nuvem.

# Criando uma lista de *stopwords*

```python
STOPWORDS = ['Página','Atualmente','sobre','atuando','do','grande','verificando','junto','Textos','Even','conhecer','cotas',
             'tive','Vínculo','Sistema','Resumo','desde','tecnicamente','concorrência','Atual','seu','ser','Idiomas',
             'Infraestrutura','Bolsista','Carga','verificação','Aplicada','Tironi','Regime','Parcial','with','sou','estou',
             'minha','Falcão','acadêmica', 'auxiliando', 'titulação','momento','Alan','Ricardo', 'Silva','vai',
             'Nacional','Pesquisa','com','de','trabalhei','Também','Server',
             'Duolingo','via','Título','trabalhando','correspondência','Por','nota',
             'as','os','Universidade','meus','aprendi','publicar','inconsistências','João','https','horária',
             'Instituto','Orientador','dos','Coursera','Fundamentos','além','que','meio','Enquadramento','funcional',
             'no','nos','como','em','Brasília','para','da','na','um','Costa','exclusiva','Parte','trabalho',
             'joaorenatofalcao','interesse','despertou','Côrtes','pt']
```

## Gerando a nuvem

```python
nuvem = WordCloud(stopwords=STOPWORDS,
                      background_color="white" , # cor de fundo da nuvem
                       width=1612, height=400)
nuvem.generate(parsed)
plt.figure( figsize=(80,40) )                    #ajustar tamanho imagem
plt.imshow(nuvem, interpolation="bilinear")
plt.axis('off')
```

Os comandos <b>width=</b> e <b>height=</b> servem para definir a largura e altura da imagem gerada, respectivamente.

Já o comando <b>background_color=</b> serve para definir a cor de fundo a ser utilizada na imagem da nuvem de palavras.

## E se eu quiser salvar a imagem gerada?

```python
nuvem.to_file('CLOUD.png')
```

# Gerando uma nuvem dentro de uma imagem 

Aqui, vamos utilizar a imagem JOAO.png para gerar a nuvem de palavras dentro da imagem.

![alt text](JOAO.png "Title")

```python
mask = np.array(Image.open("JOAO.png")) # criando o objeto mask a partir da matriz de pontos da imagem
mask
```

```python
wc = WordCloud(stopwords=STOPWORDS, 
               mask = mask,                                # basta adicionar o comando mask = para a a nuvem ser gerada 
                      background_color="white" ,           # dentro da imagem
                        scale=29,                          # o comando scale define a nitidez das palavras da nuvem
                       width=1612, height=400)
wc.generate(parsed)
plt.figure( figsize=(80,40) ) #ajustar tamanho imagem
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
```
